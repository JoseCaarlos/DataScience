```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("Classified Data", index_col=0)
```

```python
df.head()
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('TARGET CLASS',axis=1))
```

```python
df_normalizado = scaler.transform(df.drop('TARGET CLASS',axis=1))
```

```python
df_normalizado
```

```python
df_param = pd.DataFrame(df_normalizado,columns=df.columns[:-1])
```

```python
df_param
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_param,df['TARGET CLASS'],test_size=.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train,y_train)
```

```python
pred = knn.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
erro_rate = []

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred = knn.predict(x_test)
    erro_rate.append(np.mean(pred != y_test))
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(14,8))
plt.plot(range(1,40),erro_rate,color='blue',ls='dashed',marker='o')
plt.xlabel('K')
plt.ylabel('Error Rate')
```
