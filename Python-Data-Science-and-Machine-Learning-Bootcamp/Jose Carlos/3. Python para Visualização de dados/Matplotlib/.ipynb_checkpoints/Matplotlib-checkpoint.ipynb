```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x*x 
```

```python
y
```

```python
# Funcional
plt.plot(x,y,'r--')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Titulo')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
fig = plt.figure()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])
axes1.plot(x,y)
axes1.set_xlabel('Eixo x')
axes1.set_title('Titulo')
axes2.plot(y,x)
```

```python
fig,ax = plt.subplots()
ax.plot(x,x**3,'b--')
ax.plot(x,x**4,'g-.')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Titulo')
```

```python
fig,ax = plt.subplots(1,2)
```

```python
ax[0].plot(x,y,'b')
ax[0].set_title('Titulo')

ax[1].plot(x,y,'b')
ax[1].set_title('Titulo')

```

```python
fig,ax = plt.subplots(5,5)
plt.tight_layout()
```

```python
fi,axes = plt.subplots(figsize=(12,8))

axes.plot(x,y,'r-',label='x ^2')
axes.plot(y,'g-.',label='x**1/2')
axes.set_title('titulo')
axes.legend(loc=0)
```
